```python
headers = {
    'Api-Key': '89d1e5b76aa4413393430963acbbd2b1',
}

response = requests.get(link, headers=headers)
```

```python
s = response.text
j = json.loads(s)
# print(j)
```

```python
df = json_normalize(j['_embedded']['tickets']) 
# df.columns
```

```python
item_name = []
item_category = []
for x in df['_embedded.items']:
    x = str(x)
    x = x.replace("\'","\"")
    x = x.replace("None", "null")
    x = x.replace("True", "true")
    x = x.replace("False", "false")
    x = x[1 : : ]
    x = x[:-1:]
    if len(x)>0:
        y = json.loads(x)
        item_name.append(y['_embedded']['menu_item']['name'])
        item_category.append(y['_embedded']['menu_item']['_embedded']['menu_categories'][0]['name'])
    else:
        item_name.append('None')
        item_category.append('None')
    
```

```python
new_df = pd.DataFrame()
new_df['Ticket_ID'] = df['id']
new_df['Total'] = df['totals.total']
new_df['Subtotal'] = df['totals.sub_total']
new_df['Tax'] = df['totals.tax']
new_df['Tips'] = df['totals.tips']
new_df['Opened At'] = df['opened_at']
new_df['Closed At'] = df['closed_at']
new_df['Order Type'] = df['_embedded.order_type.name']
new_df['Item Name'] = item_name
new_df['Category Name'] = item_category
new_df['Closed At'] = new_df['Closed At'].fillna("None")
new_df
```

```python
new_df.set_index("Ticket_ID", inplace = True) 
```

```python
new_df.to_csv("data.csv")
```
